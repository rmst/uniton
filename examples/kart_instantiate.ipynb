```python
import os
from uniton import UnityEngine
ue = UnityEngine(os.getenv("UNITY_BUILD_PATH"))
# ue = UnityEngine()
```

```python
ue.scene.GameManager.SetActive(False)  # disable timer and GUI
ue.scene.KartClassic_Player.ArcadeKart.SetCanMove(True);
```

```python
kart = ue.scene.KartClassic_Player
```

```python
def slow(c, dt=0.05):
    from time import sleep
    for x in c:
        yield x
        sleep(dt)
```

```python
copies = [ue.GameObject.Instantiate(kart, kart.position + ue.Vector3(i, 0, j), kart.rotation) for i in range(-4, 5) for j in slow(range(4, 4*20, 4))]
```

```python
[ue.GameObject.Destroy(c) for c in slow(copies)];
```
